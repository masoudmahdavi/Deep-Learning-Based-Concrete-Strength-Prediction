<strong>Download and read CSV data with Pandas</strong>

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

<strong>Split data to label and features</strong>

```python
label = concrete_data[['Strength']]
features = concrete_data.loc[:, concrete_data.columns != 'Strength']
```

<strong>Normalize features</strong>

```python
stan = StandardScaler().fit(features)
pd.DataFrame(stan.transform(features), columns=features.columns)
```
